## Import Libraries

```python
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
%matplotlib inline
```

## EDA (Exploratory Data Analysis)

## Q1: Create a dataframe and Save that dataset inside it.

```python
df = pd.DataFrame(data=boston.data, columns=boston.feature_names)
```

## Q2: Print the head rows of the dataframe.

```python
df.head()
```

## Q3: Use histogram to show the distribution of House Prices.

```python
df['price']= boston.target
```

```python
df['price'].plot(kind="hist", bins=20, title="Histogram of price");
```

## Q4: Use a heatmap to show the correlation between features and the target labels.

```python
plt.figure(figsize=(12,10))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

## Q5: Use a lmplot to draw the relations between price and LSTAT.

```python
sns.lmplot(x="price", y="LSTAT", data=df, order=1)
```

## Q6: Use a lmplot to draw the relations between price and RM.

```python
sns.lmplot(x="price", y="RM", data=df, order=1)
```

```python
df.columns
```

## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
#data
x = df.drop(['price'], axis=1)
#target
y = df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
lr = LinearRegression()
```

## Q9: Train the Model.

```python
lr.fit(x_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(lr.intercept_) 
```

```python
coeff_df = pd.DataFrame(lr.coef_,x.columns,columns=['Coefficient'])
coeff_df
```

```python
lmpred = lr.predict(x_test)
lmpred
```

```python
Real_Values = np.array(y_test)
Real_Values
```

```python
plt.scatter(Real_Values,lmpred);
```

## Residual Histogram

```python
sns.distplot((y_test-lmpred),bins=50);
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
print('MAE:', metrics.mean_absolute_error(y_test, lmpred))
print('MSE:', metrics.mean_squared_error(y_test, lmpred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, lmpred)))
print('R^2:', r2_score(y_test, lmpred))
```
